```python
import tensorflow as tf
# this only scale the image array from 0-225 to -1 to 1
from tensorflow.keras.applications.mobilenet_v3 import preprocess_input
import tensorflow_addons as tfa
import numpy as np
import os
```

```python
model = tf.keras.models.load_model("trained_model.h5" , custom_objects={
        'F1Score': tfa.metrics.F1Score(num_classes=4, average='macro')
    })
```

```python
model.summary()
```

```python
image_path = r"test\NORMAL\NORMAL-86240-8.jpeg"
img = tf.keras.utils.load_img(image_path , target_size=(224,224))
x = tf.keras.utils.img_to_array(img)
x = np.expand_dims(x , axis=0)
x = preprocess_input(x)

```

```python
pred = model.predict(x)
```

```python
pred
```

```python
class_name = ["CNV", "DME" , "DRUSEN" , "NORMAL"]
result_index = np.argmax(pred)
print("Model Prediction : ",class_name[result_index])
```
